### Literacy Rate According to Census from [1951-2011]

```python
plt.figure(figsize=(25, 10))

# Plotting graph ðŸ“Š according to census
plt.plot(df['State&Uts'], df['Literacy_rate_1951'], marker='*', label='Literacy_rate_1951')
plt.plot(df['State&Uts'], df['Literacy_rate_1961'], marker='*', label='Literacy_rate_1961')
plt.plot(df['State&Uts'], df['Literacy_rate_1971'], marker='*', label='Literacy_rate_1971')
plt.plot(df['State&Uts'], df['Literacy_rate_1981'], marker='*', label='Literacy_rate_1981')
plt.plot(df['State&Uts'], df['Literacy_rate_1991'], marker='*', label='Literacy_rate_1991')
plt.plot(df['State&Uts'], df['Literacy_rate_2001'], marker='*', label='Literacy_rate_2001')
plt.plot(df['State&Uts'], df['Literacy_rate_2011'], marker='*', label='Literacy_rate_2011')

plt.xlabel('State&Uts',fontsize = 25)
plt.ylabel('Literacy Rate Over The Years ',fontsize = 25)
plt.title('Literacy Rate According To Census')
plt.xticks(fontsize=20,rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

## comment
The literacy rate in India has shown significant improvement over the decades from the 1951 census to the 2011 census. Here's a brief overview of how the literacy rate has evolved during this period:

1. **1951 Census**: According to the 1951 census, the overall literacy rate in India was around 18.33%. At that time, the country was still grappling with widespread illiteracy, limited access to education, and socio-economic challenges.

2. **1971 Census**: By the 1971 census, the literacy rate had increased to approximately 29.45%. This increase was partly due to various educational initiatives undertaken by the government, including the establishment of schools and adult education programs.

3. **1991 Census**: In the 1991 census, the literacy rate further improved to around 52.21%. This period saw a significant expansion of educational infrastructure, increased investment in primary education, and awareness campaigns aimed at promoting literacy.

4. **2001 Census**: By the 2001 census, the literacy rate had climbed to about 64.84%. During this period, there was a growing emphasis on universal primary education, adult literacy programs, and the adoption of policies aimed at reducing gender disparities in education.

5. **2011 Census**: According to the 2011 census, the literacy rate in India reached approximately 74.04%. This marked a significant milestone in India's journey towards achieving higher levels of literacy. Factors contributing to this improvement included continued government investment in education, expansion of educational opportunities, and greater awareness about the importance of education.

It's important to note that while there has been significant progress in improving literacy rates, challenges such as regional disparities, gender gaps, and quality of education still persist. The Indian government continues to prioritize education as a key area for development and has launched various initiatives to address these challenges and further enhance literacy rates across the country.
